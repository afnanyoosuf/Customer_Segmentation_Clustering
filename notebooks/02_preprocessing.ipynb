```python
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df.columns = ['CustomerID','Gender','Age','AnnualIncome','SpendingScore']
```

```python
df.head()
```

## Encode Gender

```python
df_encoded = df.copy()
df_encoded['Gender'] = df_encoded['Gender'].map({'Male':0,'Female':1})
df_encoded.head()
```

## Creating Features

```python
feature_1 = df_encoded[['AnnualIncome','SpendingScore']]
scaler = StandardScaler()
x1_scaler = scaler.fit_transform(feature_1)


x1_scaler[:5]
```

```python
feature_2 = df_encoded[['Age','AnnualIncome','SpendingScore']]
scaler = StandardScaler()
x2_scaaler = scaler.fit_transform(feature_2)


x2_scaaler[:5]
```
